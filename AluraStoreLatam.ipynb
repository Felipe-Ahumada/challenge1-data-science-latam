#1. Análisis de facturación



```python
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
```

```python
tienda['Precio'].isnull().any() # No existen valores nulos
```

```python
tienda['Costo de envío'].isnull().any()
```

```python
print(tienda['Precio'].sum())
```

```python
def ingresos_totales(lugar_de_venta):
    ingresos_totales_tienda = lugar_de_venta['Precio'].sum() + lugar_de_venta['Costo de envío'].sum()
    return ingresos_totales_tienda
```

```python
ingresos = [
    ingresos_totales(tienda),
    ingresos_totales(tienda2),
    ingresos_totales(tienda3),
    ingresos_totales(tienda4),
]
```

```python
plt.bar(nombres_tiendas, ingresos)
plt.title("Ingresos de cada tienda")
plt.ylabel("Ingresos (Millones)")
plt.xlabel("Tienda")
plt.ylim(1e9)
plt.grid(True)
```

La tienda con menos ingresos es la tienda 4 y la con más ingresos es la tienda 1.

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio(lugar_de_venta):
    calificacion = lugar_de_venta['Calificación'].mean()
    return round(calificacion,2)
```

```python
promedios = [
    calificacion_promedio(tienda),
    calificacion_promedio(tienda2),
    calificacion_promedio(tienda3),
    calificacion_promedio(tienda4),
]
```

```python
plt.bar(nombres_tiendas, promedios)
plt.title("Calificación promedio de cada tienda")
plt.ylabel("Calificación")
plt.xlabel("Tienda")
plt.ylim(3.94, 4.06)
plt.grid(True)
```

La peor calificada es la tienda 1 y la mejor calificada es la tienda 3.

notar que la diferencia entre la tienda 1 y la tienda 4 es de 2 centésimas. Lo cual puede indicar que no es mucha la diferencia

# 4. Productos más y menos vendidos

```python
tienda4['Producto'].isnull().sum() #Se realizó con las 4 tiendas
```

```python
def producto_mas_vendido(lugar_de_venta):
    valor_maximo = lugar_de_venta['Producto'].value_counts().max()
    cantidad_producto = lugar_de_venta['Producto'].value_counts()
    return cantidad_producto[cantidad_producto > valor_maximo-1]

def producto_menos_vendido(lugar_de_venta):
    valor_minimo = lugar_de_venta['Producto'].value_counts().min()
    cantidad_producto = lugar_de_venta['Producto'].value_counts()
    return cantidad_producto[cantidad_producto < valor_minimo+1]
```

**Tienda 1**

```python
print(f'Los producto más vendido de la tienda 1 son:\n{producto_mas_vendido(tienda).index.to_list()} \nLos productos menos vendido de la tienda 1 son: \n{producto_menos_vendido(tienda).index.to_list()}')
```

**Tienda 2**

```python
print(f'Los producto más vendido de la tienda 2 son:\n{producto_mas_vendido(tienda2).index.to_list()} \nLos productos menos vendido de la tienda 2 son: \n{producto_menos_vendido(tienda2).index.to_list()}')
```

**Tienda 3**

```python
print(f'Los producto más vendido de la tienda 3 son:\n{producto_mas_vendido(tienda3).index.to_list()} \nLos productos menos vendido de la tienda 3 son: \n{producto_menos_vendido(tienda3).index.to_list()}')
```

**Tienda 4**

```python
print(f'Los producto más vendido de la tienda 4 son:\n{producto_mas_vendido(tienda4).index.to_list()} \nLos productos menos vendido de la tienda 4 son: \n{producto_menos_vendido(tienda4).index.to_list()}')
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
etiquetas_y = [] 
valores_x = []

for nombre, df_tienda in zip(nombres_tiendas, tiendas):
    resultado = producto_mas_vendido(df_tienda)
    for producto, cantidad in resultado.items():
        etiquetas_y.append(f"{nombre}\n({producto})")
        valores_x.append(cantidad)
```

```python
plt.barh(etiquetas_y, valores_x)
plt.title('Productos Más Vendidos por Tienda')
plt.xlabel('Cantidad de Ventas')
```

```python
etiquetas_menos = []
valores_menos = []

for nombre, df_tienda in zip(nombres_tiendas, tiendas):
    resultado = producto_menos_vendido(df_tienda)
    for producto, cantidad in resultado.items():
        etiquetas_menos.append(f"{nombre}\n({producto})")
        valores_menos.append(cantidad)
```

```python
plt.barh(etiquetas_menos, valores_menos)

plt.title('Productos Menos Vendidos por Tienda')
plt.xlabel('Cantidad de Ventas')
```

# 5. Envío promedio por tienda

```python
def envio_promedio(lugar_de_venta):
    costo = lugar_de_venta['Costo de envío'].mean()
    return f'El costo promedio del envío es de: {costo:.2f}'
```

```python
print("Tienda 1: " + envio_promedio(tienda))
print("Tienda 2: " + envio_promedio(tienda2))
print("Tienda 3: " + envio_promedio(tienda3))
print("Tienda 4: " + envio_promedio(tienda4))
```

El que mayor costo de envío tiene es la tienda 1 y el de menor costo de envío es la tienda 4. Pero la tienda 1 tiene la mayor cantidad de compras por lo que puede que el gasto de envío no sea relevante (Se presupone).
